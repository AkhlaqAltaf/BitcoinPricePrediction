Data preprocessing and cleaning

drop unwanted column

```python
import pandas as pd 
df = pd.read_csv("dataset/task-1.zip")
df
```

no of null values in Close column : 1243608

```python
df.drop(columns =["Weighted_Price","Close","Timestamp"],inplace=True)
```

As NAN values are same in every column so we drop those rows

```python

df.dropna(inplace=True)
df
```

finding and showing correlation

```python
import seaborn as sbn
correlation = df.corr()
sbn.heatmap(correlation,annot=True,cmap = "coolwarm")
```
